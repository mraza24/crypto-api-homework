```python
import requests
import pandas as pd

# 1. Get all coins data
url_all = "https://api.binance.com/api/v3/ticker/price"
response_all = requests.get(url_all)

if response_all.status_code == 200:
    coins_data = response_all.json()
    df = pd.DataFrame(coins_data)
    print("✅ All Coins Data Fetched Successfully!\n")
    display(df.head(10))   # sirf first 10 show karenge
else:
    print("❌ Error fetching all coins:", response_all.status_code)


# 2. Get single coin data by symbol
symbol = "BTCUSDT"   # Example: Bitcoin in USDT
url_single = f"https://api.binance.com/api/v3/ticker/price?symbol={symbol}"
response_single = requests.get(url_single)

if response_single.status_code == 200:
    coin_data = response_single.json()
    print(f"\n✅ {symbol} Price:")
    print(coin_data)
else:
    print("❌ Error fetching single coin:", response_single.status_code)
```
